```python
# Notes & Lecture by Akhona Njeje.
# Date 23 May 2023.
# Topic : Machine Learning & Algorithms.KMC.
# Documentation info : https://en.wikipedia.org/wiki/Scikit-learn & SKLearn Cheat sheet.
# Doc & Extra read ---> Intro to Stats Learning Chapt 10 by Gareth James.

# Theory.

# KMC is a unsupervised learning Algorithm.
# KMC attempts to group similar clusters together in your data.
# Clustering problems for KMC.
   # Market segmentation.
   # Cluster similar documents.
# The overall goal is divide data into distinct groups such that observations within group are similar.
# K Means Clustering :
   # Choose a number of clusters "K".
   # Randomly assign each point a cluster. 
   # Works best with Mean Vector method.
    
    
# Import frameworks & libraries.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs   # Blobs, helps us create Artificial data for clustering.
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:,0],data[0][:,1])
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1])
```

```python
# Model Development.
# KMC Unsupervised Algorithm.

from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_   # These are the labels our model belives them to be true for our clusters.
```

```python
fig , (ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],cmap='rainbow')
```
